# IN THIS QUES THERE IS A SCHOOL WHICH 

### To compare sample mean with population mean and check whether the difference is real or due to chance.
### To test hypotheses when sample size is large usually n ≥ 30, or population standard deviation is known.

```python
import numpy as np
import scipy.stats as stats
```

provide problem statment


```python
sample_mean = 110
population_mean =100
population_standard_deviation=15
sample_size =50 
alpha = 0.05
CI = 0.95
```

test applied

```python
z_score = (sample_mean - population_mean )/ (population_standard_deviation/ np.sqrt(sample_size))
```

```python
z_score
```

calculate z_critical value

```python
#inside scipy.stats we have module to calculate  z_test
z_critical=stats.norm.ppf(CI)  # for one tailed

# two tailed 
# z_critical =stats.norm.ppf(1-(alpha/2))

```

```python
if (z_score > z_critical) :
    print("REJECT THE NULL HYPOTHESIS ")
else :
    print("WE FAIL TO REJECT THE NULL HYPOTHESIS")
```

CONCEPT OF P VALUE 

```python
p_value =1-0.999966
```

```python
if p_value > alpha :
    print("we fail to reject the null hypothesis")
else :
    print("reject the null hypothesis")
```

# Example 2

```python
from scipy import stats

before_training = [20, 25, 18, 30, 22, 28, 24, 26, 19, 21]
after_training  = [25, 30, 20, 35, 24, 32, 27, 29, 22, 25]

t_stat, p_value = stats.ttest_rel(before_training, after_training)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value <= alpha:
    print("Reject Null Hypothesis")
    print("Training made a significant difference.")
else:
    print("Fail to Reject Null Hypothesis")
    print("Training did not make a significant difference.")
```

# EXAMPLE FOR TWO INDEPENDENT 


```python
from scipy import stats

team_a_scores = [45, 50, 55, 48, 52, 60, 58, 49]
team_b_scores = [35, 40, 42, 38, 45, 41, 39, 44]

t_stat, p_value = stats.ttest_ind(team_a_scores, team_b_scores)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value <= alpha:
    print("Reject Null Hypothesis")
    print("There is a significant difference between Team A and Team B scores.")
else:
    print("Fail to Reject Null Hypothesis")
    print("There is no significant difference between Team A and Team B scores.")
```
